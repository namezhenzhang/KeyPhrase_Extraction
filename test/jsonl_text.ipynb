```python
import jsonlines
max_ = 1000000
line = 0
texts = []
with open("OpenKPEval.jsonl", "r+", encoding="utf8") as f:
    for item in jsonlines.Reader(f):
        if line < max_:

            text = {}
            text['url'] = item['url']
            text['text'] = item['text']
            text['KeyPhrases'] = item['KeyPhrases']

            texts.append(text)

            line += 1
        else:
            break
# print(len(length),max(length),min(length))
```
